# JSON File

JSON ----> Java Script Oject Notation
Json is often used when data is used to store and exchange between sever and web page.
Json is made-up of key value pairs, just like a Python dictionary, starts with `{` and ends with `}`

`{
  "name": "Alice",
  "age": 30,
  "city": "New York",
  "email": "alice@example.com"
}`

```python
import pandas as pd
import requests
from io import StringIO
```

## Working with Json

```python
url = "https://raw.githubusercontent.com/Malekai/Downloading-Data/master/population_data.json"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
json_df = StringIO(req.text)
```

```python
pd.read_json(json_df)
```

```python
df = pd.DataFrame()  # Initialize an empty DataFrame

for i in range(1, 429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    if response.status_code == 200:
        temp_data = response.json().get('results')
        if temp_data:  # Check if there's data in the response
            temp_df = pd.DataFrame(temp_data)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
            if df.empty:  # Check if the DataFrame is still empty
                df = temp_df.copy()  # If empty, copy the data into df
            else:
                df = pd.concat([df, temp_df], ignore_index=True)  # Concatenate the new data to the existing DataFrame
        else:
            print(f"No data found on page {i}")
    else:
        print(f"Failed to fetch data from page {i}")

# Now df contains all the collected movie data from multiple pages
```
