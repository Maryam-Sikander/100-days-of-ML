## Opening a csv file from URL

```python
import pandas as pd
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/asad70/reddit-sentiment-analysis/master/list%20of%20subreddits.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

data = pd.read_csv(data, chunksize=8)
```

## Working with tsv file

```python
url = "https://raw.githubusercontent.com/ccusa/Disaster_Vulnerability_Map/master/sample-data.tsv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
df_tsv = StringIO(req.text)
```
